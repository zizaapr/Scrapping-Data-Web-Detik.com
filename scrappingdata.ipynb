```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv

HEADERS = HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
```

```python
def scrape_detik(num_pages):
    data_list = []
    seen_titles = set()
    base_url = "https://www.detik.com/search/searchnews?query=pemilu+2024&sortby=time&page={}"
    index = 1

    for page in range(1, num_pages + 1):
        url = base_url.format(page)
        response = requests.get(url, headers=HEADERS)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            articles = soup.find_all("article")
            li = soup.find("div", {"class": "section nh1", "id": "nh1"})

            if not articles:
                print(f"Tidak menemukan artikel di halaman {page}")
                continue

            for article in articles:
                try:
                    link = article.find("a")["href"] if article.find("a") else "Tidak ada link"
                    headline = article.find("h3").text.strip() if article.find("h2") else "Tidak ada judul"

                    if headline in seen_titles:
                        continue
                    seen_titles.add(headline)

                    news_response = requests.get(link, headers=HEADERS)
                    news_soup = BeautifulSoup(news_response.text, "html.parser")

                    date_elem = news_soup.find("div",class_="detail__date")
                    if date_elem:
                        date = date_elem.text.strip("div").replace("WIB", "")
                    else:
                        date_elem = news_soup.find("div",class_="detail__date")
                        date = date_elem.text.strip("div").replace("WIB", "") if date_elem else ""

                    content_section = news_soup.find("div", class_="detail__body-text itp_bodycontent")
                    content = " ".join(p.text.strip() for p in content_section.find_all("p")) if content_section else "Konten tidak ditemukan"

                    data_list.append([index, headline, date, link, content])
                    print(f"done [{index}] > {headline[0:10]}")
                    index += 1

                except Exception as e:
                    print(f"Error scraping article: {e}")

        else:
            print(f"Gagal mengambil halaman {page}, status code: {req.get(url, headers=HEADERS).status_code}")

    save_to_csv(data_list)

def save_to_csv(data_list, filename="politik.csv"):
    if not data_list:
        print("Tidak ada data baru untuk disimpan.")
        return

    with open(filename, "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file, delimiter=";")
        writer.writerow(["No", "Judul", "Tanggal", "Link", "Isi Berita"])
        writer.writerows(data_list)

    print(f"Data berhasil disimpan ke {filename}")
```
